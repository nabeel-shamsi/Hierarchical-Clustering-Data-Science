```python
#Importing the Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
dataset = pd.read_csv("Mall_Customers-1.csv")
X = dataset.iloc[:, :].values
```

```python
X
```

```python
#Dendrogram to find optimum number of clusters
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(X, method = 'ward'))

#Giving title to dendrogram
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
```

```python
#now we have to train the model
from sklearn.cluster import AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters = 5)
y_hc = clustering.fit_predict(X)
```

```python
#Now Visualizating the clusters
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], c = 'red', label = 'Cluster1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], c = 'blue', label = 'Cluster2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], c = 'yellow', label = 'Cluster3')
plt.scatter(X[y_hc == 3, 0], X[y_hc == 3, 1], c = 'green', label = 'Cluster4')
plt.scatter(X[y_hc == 4, 0], X[y_hc == 4, 1], c = 'orange', label = 'Cluster5')

#Giving title to scatter plot
plt.title("Cluster of Customers")
plt.xlabel("Annual Income(K$)")
plt.ylabel("Spending Score(1-100)")
plt.legend()
plt.show()
```

```python
# let categorize them into 3 clusters
clustering = AgglomerativeClustering(n_clusters = 3)
y_hc = clustering.fit_predict(X)
```

```python
y_hc
#here we see , there are 3 types of category
```

```python
#now plot it
plt.scatter(X[y_hc == 0, 0], X[y_hc == 0, 1], c = 'red', label = 'Cluster1')
plt.scatter(X[y_hc == 1, 0], X[y_hc == 1, 1], c = 'blue', label = 'Cluster2')
plt.scatter(X[y_hc == 2, 0], X[y_hc == 2, 1], c = 'yellow', label = 'Cluster3')


plt.title("Cluster of Customers")
plt.xlabel("Annual Income(K$)")
plt.ylabel("Spending Score(1-100)")
plt.legend()
plt.show()
```
